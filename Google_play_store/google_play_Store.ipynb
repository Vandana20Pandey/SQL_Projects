### Data Preprocessing

```python
# Dropping null values
data.dropna(inplace=True)
```

```python
# Dropping unnecessary columns
data.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
# Changing datetime to desired format
data['Last Updated'] = pd.to_datetime(data['Last Updated'], errors='coerce')
```

```python
# Getting Last Updated column in 'YY-MM-DD' format
data['Last Updated'] = data['Last Updated'].dt.strftime('%y-%m-%d')
```

```python
data['Last Updated']=  pd.to_datetime(data['Last Updated'],errors='coerce')
```

```python
data.info()
```

```python
# Cleaning Install Columns
data['Installs'] = data['Installs'].str.replace('+', '')
data['Installs'] = data['Installs'].str.replace(',', '')
```

```python
# Converting Rating, Review and install columns in desired datatype
data['Rating'] = data['Rating'].astype(float)
data['Reviews'] = data['Reviews'].astype(int)
data['Installs'] = data['Installs'].astype(int)
```

```python
data.sample(10)
```

```python
# Removing Dollar sign from Price column
data['Price'] = data['Price'].str.replace('$', '')
```

```python
# Changing Datatype of price column from string to float
data['Price'] = data['Price'].astype(float)
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
# Getting the cleaned data in csv format
data.to_csv('playstore.csv', index=False)
```
